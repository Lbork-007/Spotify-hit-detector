```python
tracks_features = pd.read_csv("/content/spotify-12m-songs/tracks_features.csv")
```

```python
billboard_songs = pd.read_csv("/content/drive/MyDrive/Project/HotSongsBillBoard.csv")
```

```python
tracks_features.columns
```

```python
tracks_features_processed = tracks_features.drop(['id', 'album_id', 'artist_ids', 'track_number', 'disc_number'],axis=1)
```

```python
tracks_features_processed.head()
```

```python
str(tracks_features_processed.artists[0])
tracks_features_processed[-2:-1]
tracks_features_processed[0:1]
```

```python
tracks_features_processed['artists'] = tracks_features_processed['artists'].apply(lambda x: x[1:-1])
```

```python
tracks_features_processed.head()
```

```python
billboard_songs.head()
```

```python
tracks_features_processed['in_billboard'] = False
```

```python
x = tracks_features_processed['release_date']
x = x.str[:4].astype(int)
def check_year(value):
    return 1990 < value < 2012
result = x.apply(check_year)
```

```python
tracks_features_processed2 = tracks_features_processed[result]
```

```python
tracks_features_processed = tracks_features_processed2.reset_index()
```

```python
tracks_features_processed['id'] = tracks_features_processed.name + tracks_features_processed.artists
billboard_songs['id'] = billboard_songs.track + billboard_songs.artist
```

```python
tracks_features_processed['in_billboard'] = (tracks_features_processed['id'].isin(billboard_songs['id']))
```
