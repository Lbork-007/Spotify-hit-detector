```python
# prompt: can you convert dictionary 'track_features' into a dataframe?

import pandas as pd

def get_song_features(song_link):
    # Initialize Spotipy client
    client_id = '69e43c60759e4f3783cc9a6384e52022'
    client_secret = '3ab41d22f522477d93a427a1761ff2a0'
    client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
    sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

    # Extract track ID from the Spotify link
    track_id = song_link.split('/')[-1]

    # Get audio features for the track
    track_features = sp.audio_features(track_id)
    track_features1 = pd.DataFrame(track_features)
    track_features1 = track_features1[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']]

    # Convert dictionary to dataframe
    return track_features1
```

```python
def save_features_to_csv(features, filename='song_features.csv'):
    with open(filename, 'a', newline='') as csvfile:
        features.to_csv(csvfile)
```

```python
listofsongs = pd.read_csv('/content/drive/MyDrive/Project/2000s_pop_hit_throwback_.csv')

ids = listofsongs['Spotify ID']
names = listofsongs['Track Name']
# You can adjust this template according to your actual link structure
links = [f'https://open.spotify.com/track/{id}' for id in ids]
```

```python
song_df = pd.DataFrame()
for link in links:
  song_link = link
  song_features = get_song_features(song_link)
  if song_features.empty == False:
      ##print("Song features retrieved successfully.")
      ##print(song_features)
      ##print("Song features saved to CSV file.")
      song_df = song_df.append(song_features)
  else:
      print("Failed to retrieve song features.")
```

```python
with open('/content/drive/MyDrive/Project/rf_model.pkl', 'rb') as file:
  rf_model = pickle.load(file)
```

```python
y_pred = rf_model.predict_proba(song_df)
y_pred
```

```python
output = pd.concat([pd.DataFrame(y_pred), names], axis = 1)
```

```python
# prompt: can you sort the DataFrame "output" based on its second column in descending order?

output = output.sort_values(by=[1], ascending=False)
```
